```python
import requests
from IPython.display import Image, display
from ipywidgets import widgets, HBox
import itertools

class Pokemon:
    def __init__(self,pokename):
        self.pokemon_url=f'https://pokeapi.co/api/v2/pokemon/{pokename.lower()}'
        self.pokename=pokename.title()
        self.json=self.get_data()
        self.profile={}
        self.stats={}
        self.abilities=[]
        self.types=[]
        self.build_profile()

    
    def get_data(self):
        data=requests.get(self.pokemon_url)
        if(data.ok):
            return(data.json())
        else:
            print(f"{self.pokename} was too elusive try a different Pokemon")
            return({})
            
    def get_stats(self):
        statkeys=[]
        statvals=[]
        for stat in self.json["stats"]:
            statkeys.append(stat["stat"]["name"])
            statvals.append(stat["base_stat"])
        i=0
        for key in statkeys:
            self.stats[key]=statvals[i]
            i+=1
       
        
    def get_abil(self):
        if self.json:
        for ability in self.json["abilities"]:
            self.abilities.append(ability["ability"]["name"])
            
    def get_types(self):
        if self.json:
            for atype in self.json['types']:
                self.types.append(atype['type']['name'])
            
    def build_profile(self):
        if self.json:
            self.get_abil()
            self.get_types()
            self.get_stats()      
            self.name=f"{self.json['name']}"
            self.profile={self.name:{
                    "abilities":self.abilities,
                    "base_stats":self.stats,
                    "types":self.types,
                    "height":self.json['height'],
                    "weight":self.json['weight'],
                    "name":self.name
                }
            }


    def show_icon(self):
        if self.json:
            spriteurl=self.json['sprites']['front_shiny']
            display(Image(spriteurl))
        
    def show_icon_in_list(self):
        if self.json:
            spriteurl=self.json['sprites']['front_shiny']
            my_image = widgets.HTML(value=f'<img src={spriteurl}>')
            return(my_image)


class PokeDeck:
    def __init__(self):
        self.deck=[]
        self.pokes=[]
        self.other_deck=[]
    def collect(self, pokemon):##adds to deck Gotta Collect EM All
        if(pokemon.json):
            self.deck.append(pokemon.profile)
            self.pokes.append(pokemon)
        
    def show_deck(self):
        img_list=[]
        for poke in self.pokes:
            img_list.append(poke.show_icon_in_list())
        display(HBox(img_list))
                        
    def get_deck(self):
        return self.deck
    
    def show_deck_by_type(self):
        typegrp=[]
        newdict={}
        for poke in self.pokes:
            if poke.types[0] not in typegrp:
                typegrp.append(poke.types[0])
        i=0
        for grp in typegrp:
            newdict[grp]=[]
            i+=1
      
        for grp in typegrp:       
            for poke in self.pokes:
                if poke.types[0]==grp:
                    newdict[grp].append(poke.profile)
        self.other_dict=newdict
        for k, v in self.other_dict.items():
            print(f"{k.title()}: ")
            for dic in v:
                for i, j in dic.items():
                    print(f"    {i.title()}: ")
                   
                    for n, m in j.items():
                        print(f"        {n}",end="   :   ")
                        print(f"{m}")

        
            
```

```python
pikachu=Pokemon('pikachu')
abcd=Pokemon('abcd')
venusaur=Pokemon('Venusaur')
charmander=Pokemon('charMander')
raichu=Pokemon('Raichu')

my_deck = PokeDeck()
my_deck.collect(pikachu)
my_deck.collect(abcd)
my_deck.collect(venusaur)
my_deck.collect(charmander)
my_deck.collect(raichu)
my_deck.show_deck()
my_deck.show_deck_by_type()
```
